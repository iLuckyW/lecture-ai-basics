Support Vector Machines (SVM) beschreiben eine ML-Methode f√ºr supervised Learning.

```python
import pandas as pd

dataset = pd.read_csv("lung_cancer_dataset.csv")
#print(f"Initial dataset size: {len(dataset)} rows")

# Einfaches random sampling (ohne stratification) // reduce rows to 20%
#dataset = dataset.sample(frac=0.1, random_state=42).reset_index(drop=True)
#print(f"Dataset size after reduction: {len(dataset)} rows")

# Columns to reduce complexity of dataset: (columns with low impact on pca)
# asbestos_exposure
# secondhand_smoke_exposure
# copd_diagnosis
# family_history
###
# gender -> only male


# Convert text to categorical data
dataset['gender'] = dataset['gender'].astype('category')
dataset['radon_exposure'] = dataset['radon_exposure'].astype('category')
dataset['alcohol_consumption'] = dataset['alcohol_consumption'].fillna('None').astype('category')

dataset['asbestos_exposure'] = dataset['asbestos_exposure'].map({'Yes': True, 'No': False})
dataset['secondhand_smoke_exposure'] = dataset['secondhand_smoke_exposure'].map({'Yes': True, 'No': False})
dataset['copd_diagnosis'] = dataset['copd_diagnosis'].map({'Yes': True, 'No': False})
dataset['family_history'] = dataset['family_history'].map({'Yes': True, 'No': False})
dataset['lung_cancer'] = dataset['lung_cancer'].map({'Yes': True, 'No': False})

# no duplicate rows
dataset.duplicated().sum()

# show data
dataset.head()
```

```python
from sklearn.model_selection import train_test_split

# get the data and target from the data frame 
data = dataset.loc[:, 'age':'family_history']
target = dataset['lung_cancer']

#print(f"Initial dataset size: {len(dataset)} rows")

#X_reduced, _, y_reduced, _ = train_test_split(data, target, train_size=0.1, random_state=42, stratify=target)
train_data, test_data, train_label, test_label = train_test_split(data, target, test_size=0.2, random_state=42)

#print(f"Dataset size after reduction: {len(X_reduced)} rows")

#train_data, test_data, train_label, test_label = train_test_split(data, target, test_size=0.3, random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Spaltennamen nach Typ trennen
categorical_cols = data.select_dtypes(include=['category']).columns.tolist()
numeric_cols = data.select_dtypes(include=['number', 'bool']).columns.tolist()

# Preprocessing-Transformer (numerisch: RobustScaler, kategorisch: OneHotEncoder)
preprocessor = ColumnTransformer([
    ("numeric", "passthrough", numeric_cols), # RobustScaler() hat keine/kaum Auswirkung
    ("categorical", OneHotEncoder(drop='first'), categorical_cols)
])

X_processed = preprocessor.fit_transform(train_data)

# Get feature names from OneHotEncoder
cat_feature_names = preprocessor.named_transformers_['categorical'].get_feature_names_out(categorical_cols)
all_feature_names = numeric_cols + list(cat_feature_names)

# Create DataFrame
X_processed_df = pd.DataFrame(X_processed, columns=all_feature_names)
X_processed_df.head()
```

Allg. gilt:
- k beschreibt die Anzahl der Folds einer Cross-Validation (wie oft Datne in Trainings- und Testsets aufgeteilt werden)
    - -> beeinflusst Robustheit des Modells & nicht das Modell selbst
- C beschreibt die St√§rke der Regularisierung vom SVM-Modell
    - -> kleiner C-Wert (z.B. 0.0001) = starke Regularisierung // das Modell toleriert mehr Fehler im Training
    - -> gro√üer C-Wert (z.B. 1) = schw√§chere Regularisierung // das Modell passt sich st√§rker an die Trainingsdaten an

```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

svc_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', SVC()),
])
```

Mithilfe von GridSearchCV soll man laut CoPilot die Hyperparameter (wie z.B. C, gamma) verbessern k√∂nnen

```python
from sklearn.experimental import enable_halving_search_cv  # Ben√∂tigt f√ºr HalvingGridSearchCV
from sklearn.model_selection import HalvingGridSearchCV

# Parameter-Raster definieren
param_grid = {
    'classifier__C': [0.01, 0.1, 1, 10, 100],
    'classifier__kernel': ['linear', 'rbf', "sigmoid", "poly"],  # precomputed kann nur mit vorgegebenen Kernel-Matrizen verwendet werden # poly dauert zu lange
    'classifier__gamma': ['auto', 'scale', 0, 0.1, 0.3, 0.5, 0.7, 1],  # gamma nur f√ºr rbf/poly/sigmoid
    #'classifier__class_weight': ['balanced', None], # balanciert ungleiche Klassenverteilungen aus // nicht sinnvoll, weil es sich um einen ausgewogenen Datensatz handelt
    'classifier__degree': [2, 3, 4], # nur f√ºr poly // steuert Grad der Polynomfunktion
    'classifier__coef0': [0.0, 0.1, 0.5], # poly/sigmoid // steuert Einfluss von h√∂herwertigen vs. niederwertigen Merkmalen (Form des Kernels)
    #'classifier__shrinking': [True, False], # ob Shrinking-Algorithmus verwendet wird // unwichtige Vektoren werden entfernt, um Training zu beschleunigen
    #'classifier__probability': [True, False], # ob Wahrscheinlichkeiten gesch√§tzt werden (langsamer) // macht logitistische Regression
    #'classifier__tol': [1e-3, 1e-4] # Toleranz f√ºr die Optimierung
}

# HalvingGridSearchCV mit SVM - deutlich schneller als GridSearchCV
halving_search = HalvingGridSearchCV(
    svc_pipeline, 
    param_grid, 
    cv=5, 
    scoring='accuracy', 
    n_jobs=-1, 
    verbose=2,
    factor=3,  # Faktor f√ºr die schrittweise Reduzierung (1/3 bleiben in jeder Iteration)
    min_resources='smallest',  # Automatische Bestimmung der minimalen Ressourcen
    max_resources='auto',  # Automatische Bestimmung der maximalen Ressourcen
    aggressive_elimination=False  # Konservativere Elimination f√ºr bessere Ergebnisse
) 
# HalvingGridSearchCV startet mit wenigen Samples/Kandidaten und steigert sich schrittweise
# Nur die besten Parameter-Kombinationen werden mit mehr Ressourcen getestet
halving_search.fit(train_data, train_label)

# Beste Parameter und Modell
print("Beste Parameter:", halving_search.best_params_)
best_model = halving_search.best_estimator_
```

```python
print("Bestes Modell:", best_model)
print("Beste Parameter:", halving_search.best_params_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ergebnisse als DataFrame
results = pd.DataFrame(halving_search.cv_results_)

# Beispiel: Vergleich der Kernels bei verschiedenen C-Werten
kernels = results['param_classifier__kernel'].unique()
c_values = results['param_classifier__C'].unique()

plt.figure(figsize=(10, 7))
for kernel in kernels:
    mask = results['param_classifier__kernel'] == kernel
    plt.plot(
        results[mask]['param_classifier__C'],
        results[mask]['mean_test_score'],
        marker='o',
        label=f'Kernel: {kernel}'
    )

plt.xlabel('C')
plt.ylabel('Mean CV Accuracy')
plt.title('HalvingGridSearchCV: Vergleich der SVM-Kernel')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, precision_score, recall_score

y_pred_train = best_model.predict(train_data)
y_pred_test = best_model.predict(test_data)

print("Accuracy Trainingsdaten:", accuracy_score(train_label, y_pred_train))
print("F1-Score Trainingsdaten:", f1_score(train_label, y_pred_train))
print("Precision Trainingsdaten:", precision_score(train_label, y_pred_train))
print("Recall Trainingsdaten:", recall_score(train_label, y_pred_train))

print("Accuracy Testdaten:", accuracy_score(test_label, y_pred_test))
print("F1-Score Testdaten:", f1_score(test_label, y_pred_test))
print("Precision Testdaten:", precision_score(test_label, y_pred_test))
print("Recall Testdaten:", recall_score(test_label, y_pred_test))

print("Confusion Matrix:\n", confusion_matrix(test_label, y_pred_test))
```

```python
from sklearn.metrics import classification_report

print(classification_report(train_label, y_pred_train))

print(classification_report(test_label, y_pred_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize=(10, 8))
_ = ConfusionMatrixDisplay.from_estimator(
    best_model, 
    test_data, 
    test_label, 
    display_labels=dataset['lung_cancer'].unique(),
    cmap=plt.cm.Blues,
    ax=ax
)
```

```python
from sklearn.model_selection import validation_curve

# collect data for plotting
train_scores, valid_scores = validation_curve(
    best_model, 
    X=train_data, 
    y=train_label, 
    param_name='classifier__C', 
    param_range=param_grid['classifier__C'], 
    cv=5
)

# first, calculate the means and standard deviations
train_scores_mean = train_scores.mean(axis=1)
train_scores_std = train_scores.std(axis=1)
valid_scores_mean = valid_scores.mean(axis=1)
valid_scores_std = valid_scores.std(axis=1)

# create the figure
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(c_values, train_scores_mean, label="Training score", color="darkorange")
ax.fill_between(
    c_values,
    train_scores_mean - train_scores_std,
    train_scores_mean + train_scores_std,
    color="darkorange",
    alpha=0.2
)

ax.plot(c_values, valid_scores_mean, label="Cross-validation score", color="navy")
ax.fill_between(
    c_values,
    valid_scores_mean - valid_scores_std,
    valid_scores_mean + valid_scores_std,
    color="navy",
    alpha=0.2
)

ax.set(
    title=f"Validation Curve with best model (Kernel: {best_model.named_steps['classifier'].kernel})",
    xlabel="C",
    ylabel="Score"
)
ax.legend(loc="best")
fig.show()
```

### 3.5 Support Vector Classifier (SVC) ‚Äì Ergebnisse und Einordnung

Im Rahmen der SVM-Analyse wurde eine Pipeline aus Vorverarbeitung (numerische Merkmale im Durchsatz, kategoriale Merkmale via One-Hot-Encoding) und einem SVC verwendet. Die Hyperparameter-Optimierung erfolgte mit HalvingGridSearchCV, wodurch die vielversprechendsten Parameterkombinationen iterativ fokussierter evaluiert wurden. Als Ergebnis liefert `best_params_` die optimale Parametrisierung (z. B. C, Kernel, gamma), w√§hrend `best_estimator_` das damit neu gefittete, direkt einsatzf√§hige Pipeline-Modell bereitstellt.

- G√ºtema√üe: Auf dem Testset wurden Accuracy und F1-Score ausgewertet und zus√§tzlich eine Confusion Matrix erstellt. Diese erlaubt eine differenzierte Betrachtung der beiden Klassen und zeigt, in welchen Bereichen Fehlklassifikationen auftreten.
- Klassenungleichgewicht: Da der Datensatz einen hohen Anteil an ‚ÄûKrebs‚Äú-F√§llen aufweist, ist die reine Accuracy nur eingeschr√§nkt aussagekr√§ftig. Der F1-Score sowie Precision/Recall pro Klasse (insbesondere Recall der Krebs-Klasse) sind daher zentrale Gr√∂√üen f√ºr die Beurteilung.
- Validierung der Regularisierung: √úber eine Validierungskurve in Abh√§ngigkeit von C wurde das Bias-Variance-Verhalten untersucht. Niedrige C-Werte (starke Regularisierung) f√ºhren eher zu h√∂herem Bias, w√§hrend hohe C-Werte das Risiko von Overfitting erh√∂hen. Die Kurve zeigt einen Bereich, in dem der Cross-Validation-Score stabil bleibt ‚Äì hier findet das Modell eine gute Balance zwischen Generalisierung und Anpassung.

Interpretation
- Wenn der Recall f√ºr ‚ÄûKrebs‚Äú hoch ist und die Precision moderat, deutet das auf wenige False-Negatives bei einigen zus√§tzlichen False-Positives hin ‚Äì im medizinischen Kontext ist dies oft w√ºnschenswert, da √ºbersehene F√§lle gravierender sind als Fehlalarme.
- Eine deutliche L√ºcke zwischen Trainings- und Test-Score w√ºrde auf Overfitting hinweisen; √§hnliche Werte sprechen f√ºr ein robustes Modell. Die Validierungskurve liefert hier eine zus√§tzliche Plausibilisierung der gew√§hlten C-Einstellung.

Praktische Nutzung
- F√ºr Vorhersagen auf neuen Daten sollte `best_estimator_` verwendet werden, da es die Preprocessing-Schritte integriert und mit den besten Parametern erneut auf allen Trainingsdaten gefittet ist.
- F√ºr Transparenz lassen sich die finalen Hyperparameter mit `best_params_` dokumentieren, w√§hrend sich aus der Confusion Matrix und dem Klassifikationsbericht konkrete Fehlertypen (False-Positives/-Negatives) ablesen lassen.

In Summe zeigt der SVC in diesem Setup eine solide Leistungsf√§higkeit, insbesondere wenn die Bewertung metrikgerecht (Precision/Recall/F1) erfolgt und die Regularisierung (C) √ºber die Validierungskurve kalibriert wird. Dies erm√∂glicht eine gezielte Abstimmung zwischen Sensitivit√§t f√ºr die Krebs-Klasse und der Rate an Fehlalarmen.
